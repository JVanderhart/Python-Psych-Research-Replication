```
def reverseScoring(df, high, cols):
    '''Reverse scores on given columns
     df = your data frame,
     high = 1+highest score available
     cols = the columns you want reversed in list form'''
    df[cols] = high - df[cols]
    return df

    # The way some questions are written means that the scores have to be reversed. An example of this could be 
    # "I rarely feel anxious" the participant rates themself on a Likert scale from 1 to 5 this would need to be
    # reversed to represent a positive metric for neuroticism.
```

```
cols = ['E2','E4','E6','E8','E10','N2','N4','A1','A3','A5','A7','C2','C6','C8','O2','O4','O7']

reverseScoring(df, 6, cols)

# Runs the previously created funciton reversing the scores that are coded in reverse. These columns each represent
# a question like the example in the previous comment.
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


x_var = dc_df[['age','engnat_0','engnat_1','engnat_2','gender_0','gender_1','gender_3']]
x_var2 = dc_df[['age','gender_1']]
x_var.head(5)

y_1 = dc_df[['extra']]
y_2 = dc_df[['neuro']]


X_train, X_test, y_train, y_test = train_test_split(x_var2, y_2, test_size = .3, random_state=42)

#   While I was testing out code I came across a data scientist convention in python
#   machine learning that is to split and train models when building them. I decided 
#   to try it out here, even though it's a bit silly to do it in this particutalr use
#   case. We aren't trying to train a model to predict personality by age and
#   gender. The reported effect would be far too weak for it to predict accurately. 
#   Instead we want to know if the weak effect is significant. 

#   Implementing this code is just a way for me to play around and showcase a test
#   train split in my code. 
```

```
lm = linear_model.LinearRegression()
model = lm.fit(x_var2, y_2)

predictions = lm.predict(x_var2)
print(predictions)

#   Seeing what it would output for predicitons.
```

```
# Fit the regressor to the training data
lm.fit(X_train, y_train)

# Predict on the test data: y_pred
y_pred = lm.predict(X_test)

# Compute and print R^2 and RMSE
print("R^2: {}".format(lm.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: {}".format(rmse))

# The R^2 here is reported at 0.024, which seems to not be particularly powerful
# as a predictor of personality, this is what we would expect. 
```

```
from statsmodels.formula.api import ols
```

```
dta = dc_df
dta['gen_age'] = dc_df['gender_1']*dc_df['age']

#   Creating a new dataset for the statistical tests, as well as an interaction variable to control for
#   colinearity. Makes the coefficients non-interpretable.
```

```

formula = 'neuro ~ gender_1 + age + gen_age'
results = ols(formula, dta).fit()
hypotheses = '(gender_1 = 0)'
f_test = results.f_test(hypotheses)
print(f_test)

#   This is the part of the project I'm least familiar with becuase of the different
#   libraries in python. I've imported ols from statsmodels to test the hypothesis
#   that the effect of gender_1 (our variable for female) plus age is equal to 0.
#   The f test resturns a p value of 4.847e-124 which gives us statistical validation
#   of the effect of gender on neuroticism, although as we saw before it is quite small,
#   and therefore isn't particularly powerful as a predictor.

#   As a side note, the statistical analysis I did in the R project is more thorough
#   than the one presented here, this is because I was more interested in getting to
#   know different libraries in python for manipulating data, rather than analyzing it.
```
